Linear Regression:

Y=wX+b,

Y==> Independent Variable,
X==> Dependent Variable,

w==> weight/slope,
b==> bias/intercept

Gradient Descent:

w=w-a*dw,
b=b-a*db,
a==>learning rate,
dw==>derivative of weight,
db==>derivative of bias
